# 1389. 케빈 베이컨의 6단계 법칙 (20분)

```python
from collections import deque

# 입력값 받기
n, m = map(int, input().split())
graph = [[] for _ in range(n+1)]
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

bacon = [0 for _ in range(n+1)]


# BFS * N
for i in range(1, n+1):
    queue = deque()
    queue.append((i,0))

    visited = [0 for _ in range(n+1)]
    while queue:
        temp, ban = queue.popleft()
        if visited[temp] == 0:
            bacon[i] += ban
            ban += 1
            visited[temp] = 1
            for node in graph[temp]:
                queue.append([node, ban])

# 최소 베이컨을 갖는 노드 구하기 
minBacon = min(bacon[1:])
for i in range(1, n+1):
    if minBacon == bacon[i]:
        print(i)
        break
```

# 7562. 나이트의 이동

```python
from collections import deque

testCase = int(input())
for _ in range(testCase):
    move = [(1,2), (2,1), (1,-2), (2,-1), (-1,2), (-2,1), (-1,-2), (-2,-1)]
    l = int(input())
    X, Y = map(int, input().split())
    lastX, lastY = map(int, input().split())

    queue = deque()
    queue.append([X, Y, 0])
    visited = [[0] * l for _ in range(l)]
    while queue:
        x, y, temp = queue.popleft()
        if visited[x][y] == 0:
            visited[x][y] = 1
            if x==lastX and y==lastY:
                print(temp)
                break
            temp += 1

            for a, b in move:
                if x+a >= 0 and x+a < l and y+b >= 0 and y+b < l:
                    queue.append([x+a, y+b, temp])
```

# 1012. 유기농 배추 (30분)

```python
test_case = int(input())
for _ in range(test_case):
    m, n, k = map(int, input().split())
    array = [[0]*n for _ in range(m)]
    arrayList = []
    for _ in range(k):
        a, b = map(int, input().split())
        array[a][b] = 1
        arrayList.append([a,b])
    move = [(1,0), (-1,0), (0,1), (0,-1)]
    cnt = 0
    for x, y in arrayList:
        queue = []
        if array[x][y] == 0:
            continue
        queue.append((x,y))
        cnt+=1
        while queue:
            tempX, tempY = queue.pop()
            array[tempX][tempY] = 0
            for a, b in move:
                nextx, nexty = tempX+a, tempY+b
                if nextx>=0 and nextx<m and nexty>=0 and nexty<n and array[nextx][nexty] == 1:
                    queue.append([nextx, nexty]);
    print(cnt)
```
